```python
import pandas as pd
import numpy as np
```

1. Nested List

```python
df = pd.DataFrame([['Banda Aceh', 'Aceh', 60000],
                   ['Langsa', 'Aceh', 800],
                   ['Denpasar', 'Bali', 50000],
                   ['Pangkalpinang', 'Bangka Belitung', 60000],
                   ['Yogyakarta', 'Daerah Istimewa Yogyakarta', 50500],
                   ['Batam', 'Kepulauan Riau', 5000],
                   ['Tanjung Pinang', 'Kepulauan Riau', 4000],
                   ['Bekasi', 'Jawa Barat', 65000],
                   ['Bogor', 'Jawa Barat', 5000]],
                   columns=['Kota', 'Provinsi', 'Jumlah_peduduk']
                   )
df
```

2. Numpy

```python
df = pd.DataFrame(np.array([['Banda Aceh', 'Aceh', 60000],
                   ['Langsa', 'Aceh', 800],
                   ['Denpasar', 'Bali', 50000],
                   ['Pangkalpinang', 'Bangka Belitung', 60000],
                   ['Yogyakarta', 'Daerah Istimewa Yogyakarta', 50500],
                   ['Batam', 'Kepulauan Riau', 5000],
                   ['Tanjung Pinang', 'Kepulauan Riau', 4000],
                   ['Bekasi', 'Jawa Barat', 65000],
                   ['Bogor', 'Jawa Barat', 5000]]),
                   columns=['Kota', 'Provinsi', 'Jumlah_peduduk']
                   )
df
```

3. Dictionary

```python
df = pd.DataFrame({
    "Kota" : ['Banda Aceh','Langsa','Denpasar','Pangkalpinang','Yogyakarta','Batam','Tanjung Pinang','Bekasi','Bogor'],
    "Provinsi" : ['Aceh','Aceh','Bali','Bangka Belitung','Daerah Istimewa Yogyakarta','Kepulauan Riau','Kepulauan Riau','Jawa Barat','Jawa Barat'],
    "Jumlah_penduduk" : [60000, 800,50000,60000,50500,5000,4000,65000,5000]
})
df
```

4. Import Documen (xlsx, csv, json, dkk)

```python
# Export Dataframe yang sudah dibuat
df.to_csv('Jumlah_penduduk.csv', index=False)
```

```python
# import Dataframe
df = pd.read_csv('Jumlah_penduduk.csv')
df
```
